# 결과값을 범주형 변수로 변환

결과값(수확기 가격)을 가격대별로 범주형 변수로 변환

```python
df4 = pd.read_csv("allfeature2.csv", encoding="euc-kr", index_col="year")
```

```python
df4
```

```python
df4["수확기"]
```

```python
tnghkrrl = df4["수확기"]
tnghkr2 = []

for a in tnghkrrl:
    if a < 35000:
        tnghkr2.append(1)
        
    if 35000 <= a < 40000 :
        tnghkr2.append(2)
        
    if a >= 40000:
        tnghkr2.append(3)
        
tnghkr2
```

```python
len(tnghkrrl), len(tnghkr2)
```

```python
df4["범주"] = tnghkr2
```

## 공분산행렬 방법 사용

```python
#피쳐들의 방향 변환
feat = xdata.T
```

```python
#공분산 행렬 구하기
covmat = np.cov(feat)
```

```python
covmat
```

```python
#eigen vlaue와 eigen vector 구하기
eig_vals, eig_vecs = np.linalg.eig(covmat)
```

```python
eig_vals
```

```python
eig_vecs
```

```python
#eigen value의 최대값과 
#전체 중 비율(정보 유실 없이 가져올 수 있는 비율) 구하기
eig_vals[0]/sum(eig_vals)
```

```python
#pca로 프로젝션된 값 구하기
projectX = xdata.dot(eig_vecs.T[0])
```

```python
projectX
```

```python
#구해진 값과 원래 값을 활용한 데이터프레임 만들기
d = pd.DataFrame(projectX, columns=["pc1"])
d["y"] = 0.0
d["result"] = Y
d["ylog"] = df4.iloc
```

```python
d["ylog"] = df5.iloc[:, -1]
```

```python
#공분산행렬을 통해 구한 값을 시각화
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
sns.lmplot("pc1", "y", data = d, fit_reg=False, hue = "result")
```
