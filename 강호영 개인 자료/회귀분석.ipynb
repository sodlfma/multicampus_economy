```python
df3 = pd.read_csv("data58.csv", encoding="euc-kr", index_col="year")
```

# 데이터 표준화

```python
lencol = len(df3.columns) -3
for i in range(lencol):
    l1 = df3.iloc[:, i]
    l1max = np.max(l1)
    l1min = np.min(l1)
    lenl1 = len(l1)
    
    l2 = []
    
    for a in l1:
        x = (a-l1min)/l1max
        l2.append(x)
        
    df3.iloc[:, i] = l2
```
